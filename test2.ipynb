```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dt.dtypes
```

```python
dt['Total Cases'] = pd.to_numeric(dt['Total Cases'], errors='coerce', downcast='float')
dt['Total Deaths'] = pd.to_numeric(dt['Total Deaths'], errors='coerce', downcast='float')
dt['Total Recovered'] = pd.to_numeric(dt['Total Recovered'], errors='coerce', downcast='float')
dt['Active Cases'] = pd.to_numeric(dt['Active Cases'], errors='coerce', downcast='float')
dt['Total Test'] = pd.to_numeric(dt['Total Test'], errors='coerce', downcast='float')
dt['Population'] = pd.to_numeric(dt['Population'], errors='coerce', downcast='float')
```
